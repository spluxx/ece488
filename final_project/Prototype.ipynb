```python
import numpy as np
from skimage import io, color
from sklearn.cluster import KMeans
import math
```

```python
rgb = io.imread("images/turtle.jpeg")
```

```python
lab = color.rgb2lab(rgb, illuminant='D65')
```

```python
lab_vec = np.reshape(lab, (math.floor(lab.size/3), 3))
```

```python
kmeans = KMeans(n_clusters=24).fit(lab_vec)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
prediction = kmeans.predict(lab_vec)
```

```python
prediction_lab_vec = kmeans.cluster_centers_[prediction]
```

```python
prediction_lab = np.reshape(prediction_lab_vec, lab.shape)
```

```python
prediction_rgb = color.lab2rgb(prediction_lab, illuminant='D65')
```

```python
io.imshow(prediction_rgb)
```

```python
def transform(rgb):
    lab = color.rgb2lab(rgb, illuminant='D65')
    lab_vec = np.reshape(lab, (math.floor(lab.size/3), 3))
    kmeans = KMeans(n_clusters=24).fit(lab_vec)
    prediction = kmeans.predict(lab_vec)
    prediction_lab_vec = kmeans.cluster_centers_[prediction]
    prediction_lab = np.reshape(prediction_lab_vec, lab.shape)
    prediction_rgb = color.lab2rgb(prediction_lab, illuminant='D65')
    return prediction_rgb
```

```python
kangaroo = io.imread("images/kangaroo.jpg")
```

```python
kangaroo_transformed = transform(kangaroo)
```

```python
io.imshow(kangaroo_transformed)
```
