```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
S1 = np.array(cv2.imread("cameraman.png", cv2.IMREAD_GRAYSCALE))
```

```python
sigma = 25
```

```python
N1 = np.random.normal(0, sigma, (len(S1), len(S1[0])))
```

```python
I1 = S1 + N1
```

```python
S2 = S1/2
```

```python
N2 = np.random.normal(0, sigma, (len(S1), len(S1[0])))
```

```python
I2 = S2 + N2
```

```python
def energy(img):
    return sum(sum(img**2.0))
```

```python
energy(S1)
```

```python
energy(S2)
```

```python
energy(N1)
```

```python
energy(N2)
```

```python
energy(I1)
```

```python
energy(I2)
```

```python
energy(S1 * 3.0)
```

```python
M = S1.size
```

```python
c = np.sqrt(4*(energy(S1) + M*sigma*sigma)/(energy(S1) + 4 * M * sigma * sigma))
```

```python
c
```

```python
I2hat = I2*c
```

```python
energy(I2hat)
```

```python
SNRI1 = energy(S1)/energy(N1)
```

```python
SNRI2 = energy(S2)/energy(N2)
```

```python
SNRI2 = energy(S2*c)/energy(N2*c)
```

```python
def rmse(I, S):
    return np.sqrt(sum(sum((I-S)**2.0))/M)
```
